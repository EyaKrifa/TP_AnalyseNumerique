```python
def horner(p, x):
    if len(p) == 1:
        return p[0]
    p[-2] += x * p[-1]
    return horner(p[:-1], x)
```

```python
horner([1, -6, 15, -20, 15, -6, 1], 0.998) # x**6 - 6*x**5 + 15*x**4 - 20*x**3 + 15*x**2 - 6*x + 1 
```

**2)un programme dicho.m qui permettra de trouver les solutions de l'√©quation f(x)=0 avec eps=0.001.**

```python
def dicho(a,b,f,n):
    for i in range(n):
        c = (a+b)/2
        if f(a)*f(c) <= 0:
            b = c
        else:
            a = c
    return b
```

```python
x3=dicho(-2,-1.5,f,10)
x10=dicho(-2,-1.5,f,10)
x20=dicho(-2,-1.5,f,20)
print(x3,x10,x20)
print(x3)
```

**3)Modification du programme dicho.m pour qu'il donne le nombre d'it√©ration n√©cessaire pour avoir la solution √† $eps$ pr√®s**

```python
def dichow(a,b,f,eps):
    cp=0
    while (b-a)>eps:
        m=(a+b)/2
        if f(m)==0:
            return m
        if f(a)*f(m)<0 : 
            b=m
        else :
            a=m
        cp+=1
    return m,cp
```

```python
f=lambda x: x**3+x**2-3*x-3
dichow(-2,-1.5,f,0.0001)
```

# Exercice4:Methode Newton

# *1)*

```python
def newton(f,fprime,a,eps):
    x1=a
    x2=x1-(f(x1)/fprime(x1))
    while abs(x1-x2)>eps :
        x1=x2
        x2=x1-(f(x1)/fprime(x1))
        m=x2
    return m
```

**On consid√®re la fonction $f(x)=e^{-x}-x$. resultat de l'algorithme `newton.m` √† la >fonction $f$, en prenant $x0=0$, $eps=0.001$ **

```python
import math
from sympy import *
from sympy.abc import x

f=lambda x: math.exp(-x)-x
fprime=lambda x: -math.exp(-x)-1
newton(f,fprime,0,0.001)
```
